# 1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to keep all customers who have done transactions with us and select the join type accordingly.

```python
join1=pd.merge(Customers,Transactions,left_on='customer_Id',right_on='cust_id',how='right')
join1
```

```python
Customer_Final=pd.merge(join1,prod_cat_info,left_on='prod_cat_code',right_on='prod_cat_code',how='left')
Customer_Final
```

# 11. For all customers aged between 25 - 35, find out:
# a. What was the total amount spent for “Electronics” and “Books” product categories?


```python
dob=pd.Timestamp.today()-pd.to_datetime(Customer_Final.DOB,format='%d-%m-%Y')
age=round(((dob.dt.days)/365.25),0)
age
```

```python
Customer_Final.loc[:,'AgeGroup']=pd.cut(Customer_Final.Age,bins=[25,35,45,55,65,75],labels=['G1','G2','G3','G4','G5'])
Customer_Final.head(5)
```

```python
round((Customer_Final.loc[(Customer_Final.prod_cat.isin(['Electronics','Books']))& (Customer_Final.AgeGroup=='G1')].groupby('prod_cat')['total_amt'].sum()),2)
```
